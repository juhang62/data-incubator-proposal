This dataset is from a survey in which 2218 respondents were given a list of 21 Scotch whisky brands and
asked to indicate whether or not they drink these brands on a regular basis. We first have a glance of the count of consumptions for each brand. The barblot below basically tells us the general popularity of those brands. But there's much more awaiting to be told by the data. The next step is to have a deep dive into the data and exploit its hidden structure. 

```r
library(repr)
options(repr.plot.width=10, repr.plot.height=8)
scotch <- read.csv("scotch.csv")
cnt=colSums(scotch)
par(mai=c(1,2.5,1,1))
barplot(cnt,horiz = TRUE,las=1,main="count of consumpations")
```
